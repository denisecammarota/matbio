En este ejercicio, nos piden analizar el modelo epidemiológico SIRS, dado por el sistema dinámico:


\begin{align}
\frac{ds}{dt} &= -\beta s i + \frac{1}{\tau_{R}} r \tag{1}, \\
\frac{di}{dt} &= \beta s i - \frac{1}{\tau_{I}} i \tag{2}, \\
\frac{dr}{dt} &=  \frac{1}{\tau_{I}} i - \frac{1}{\tau_{R}} r \tag{3}, \\
\end{align}

con la condición de que la población sea constante, dada por la ecuación:

\begin{align}
s + i + r &= 1 \tag{4}. \\
\end{align}

Es decir, este modelo es un modelo sin demografía, en el cual tenemos suceptibles, infectados y recuperados, donde la inmunidad no es para toda la vida. La tasa a las que los individuos infectados se recuperan se considera constante. Lo mismo se asume para la tasa a la que los individuos recuperados se vuelven nuevamente suceptibles. ## Búsqueda de equilibrios

Primero, nos piden buscar los equilibrios del sistema. Para ello, siguiendo el análisis de sistemas dinámicos, planteamos ecuaciones siguientes, donde $(s^{*},i^{*},r^{*})$ corresponde a un punto de equilibrio. 

\begin{align}
\frac{ds}{dt} (s^{*},i^{*},r^{*}) &= 0 \tag{5}, \\
\frac{di}{dt} (s^{*},i^{*},r^{*}) &= 0 \tag{6}, \\
\frac{dr}{dt} (s^{*},i^{*},r^{*}) &= 0 \tag{7}. \\
\end{align}

De la ecuación (5), obtenemos que:

\begin{align}
\beta s^{*} i^{*} &= \frac{r^{*}}{\tau_{R}} \tag{8}. \\
\end{align}

Utilizando esta expresión en la ecuación (6), tenemos entonces que:

\begin{align}
\frac{i^{*}}{\tau_{I}}  &= \frac{r^{*}}{\tau_{R}} \tag{9}. \\
\end{align}


Esta identidad garantiza el cumplimiento de la ecuación (7), como es de esperar por la condición que vincula a $s,i$ con $r$. Finalmente, volviendo a la ecuación (6), despejamos $s^{*}$ como:

\begin{align}
s^{*}  &= \frac{1}{\beta \tau_{I}} \tag{10}. \\
\end{align}

donde hemos descartado el caso $i^{*} = 0$, ya que da origen al punto de equilibrio trivial $(1,0,0)$, donde nunca hay infectados y, por lo tanto, toda la población permanece suceptible.  

Con lo cual, hemos demostrado lo que nos pedían en el enunciado. Finalmente, utilizando la ecuación (4), podemos escribir $i^{*}$ en función de los parámetros del sistema como:

\begin{align}
i^{*}  &= \frac{\beta \tau_{I} - 1}{\beta (\tau_{I} + \tau_{R})} \tag{11}. \\
\end{align}

Esto nos resultará útil a la hora de hacer el análisis de estabilidad lineal. 


## Análisis de estabilidad lineal

Ahora, nos piden demostrar que este sistema da lugar a oscilaciones siempre amortiguadas. Para ello, vamos a realizar el análisis de estabilidad lineal correspondiente a este sistema dinámico. Como primer paso, vamos a escribir el sistema en función de dos variables independientes, gracias a la condición que vincula a las cantidades $s,i$ con $r$. Elegiremos $r$ como la variable dependiente y $s,i$ como variables independientes. Simplemente usando la ecuación (4) en la ecuación (1), tenemos que las primeras dos ecuaciones se pueden escribir como:

\begin{align}
\frac{ds}{dt} &= -\beta s i + \frac{(1 - s - i)}{\tau_{R}} r  = f(s,i) \tag{1}, \\
\frac{di}{dt} &= \beta s i - \frac{1}{\tau_{I}} i   = g(s,i)  \tag{2}
\\
\end{align}

Entonces, realizamos las derivadas de estas funciones en función de $s,i$ y las evaluamos en el punto $(s^{*},i^{*})$. Entonces, resulta que:

\begin{align}
\frac{df}{ds}(s^{*},i^{*}) &= -\beta i^{*} - \frac{1}{\tau_{R}}  , \tag{12}\\
\frac{df}{ds}(s^{*},i^{*}) &= -\beta s^{*} - \frac{1}{\tau_{R}}, \tag{13}\\
\frac{df}{ds}(s^{*},i^{*}) &= \beta i^{*}, \tag{14}\\
\frac{df}{ds}(s^{*},i^{*}) &= \beta s^{*} - \frac{1}{\tau_{I}} \tag{15}.\\
\end{align}

Después de este proceso, podemos escribir el jacobiano:

\begin{align}
J = 
\begin{pmatrix}
-\beta i^{*} - \frac{1}{\tau_{R}} & -\beta s^{*} - \frac{1}{\tau_{R}} \\
\beta i^{*} & \beta s^{*} - \frac{1}{\tau_{I}} \\
\end{pmatrix}
 \tag{16},
\end{align}

cuyos autovalores $\lambda_{1,2}$ vamos a encontrar resolviendo la ecuación característica:

\begin{align}
det(J - \lambda Id) &= 0 \tag{17}, \\ 
\end{align}

donde $Id$ es la matriz identidad de $2x2$. Calculando este determinante y realizando algo de álgebra, podemos fácilmente ver que la ecuación (16) es equivalente a la ecuación siguiente:

\begin{align}
\lambda^{2} + \lambda \Big(\beta i^{*} + \frac{1}{\tau_{R}} - \beta s^{*} + \frac{1}{\tau_{I}}\Big) + \Big( \frac{\beta i^{*}}{\tau_{I}} - \frac{\beta s^{*}}{\tau_{R}} + \frac{1}{\tau_{I} \tau_{R}} + \frac{\beta i^{*}}{\tau_{R}}   \Big) &= 0 \tag{18}, \\ 
\end{align}

Para seguir, utilizamos las expresiones del punto de equilibrio $(s^(*),i^(*))$ en función de los parámetros del sistemas. Estas expresiones las encontramos en el apartado anterior. De esta manera, con un poco de álgebra, la ecuación (18) se torna en la ecuación (19):

\begin{align}
\lambda^{2} + \lambda \Big( \frac{\beta \tau_{I} - 1}{\tau_{I} + \tau_{R}} + \frac{1}{\tau_{R}} \Big) + \Big( \frac{\beta \tau_{I} - 1}{\tau_{I} \tau_{R}}   \Big) &= 0 \tag{19}, \\ 
\end{align}

Esta es una ecuación cuadrática de segundo grado que se resuelve fácilmente. Sus dos soluciones, que llamamos $\lambda_{1,2}$ son:

\begin{align} 
\lambda_{1,2} &= -\frac{1}{2} \Big( \frac{\beta \tau_{I} - 1}{\tau_{I} + \tau_{R}} + \frac{1}{\tau_{R}} \Big) \pm \sqrt{\Big( \frac{\beta \tau_{I} - 1}{\tau_{I} + \tau_{R}} + \frac{1}{\tau_{R}} \Big)^2 - 4 \Big(\frac{\beta \tau_{I} - 1}{\tau_{I} \tau_{R}} } \Big) \tag{20}, \\ 
\end{align}

y corresponden a dos autovalores conjugados complejos de parte real negativa. Por lo tanto, este equilibrio corresponde a una espiral estable y demostramos la presencia de oscilaciones amortiguadas. El hecho de que la parte real de estos autovalores es negativa se ve fácilmente, dado que, para que esto suceda, debe cumplirse que: 

\begin{align} 
\Big( \frac{\beta \tau_{I} - 1}{\tau_{I} + \tau_{R}} + \frac{1}{\tau_{R}} \Big) &> 0 \tag{21}. \\ 
\end{align}

Tras un poco de álgebra, se desprende que esta condición se cumple si y sólo si:

\begin{align} 
 \beta \tau_{R} &> -1 \tag{22}, \\ 
\end{align}

la cual se satisface trivialmente porque tanto $\beta$ como $\tau_{R}$ son números reales positivos. 


## Diagrama de fases y trayectorias (computacionalmente, con Python)

```python
#importo librerías que me van a ser útiles en hacer las simulaciones
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
plt.style.use('seaborn')
```

```python
# ecuaciones del SIRS 
def deriv(y, t, beta, tauI,tauR):
    S, I, R = y
    dSdt = -beta * S * I + R/tauR
    dIdt = beta * S * I  -  I/tauI
    dRdt =  I/tauI - R/tauR
    return dSdt, dIdt, dRdt
```

Primero, podemos graficar el diagrama de fases en torno al punto de equilibrio, como una manera de constatar computacionalmente que el resultado del análisis lineal es correcto. Allí, dibujaremos también algunas trayectorias del sistema para diferentes condiciones iniciales arbitrarias. Adicionalmente, señalamos el punto de equilibrio correspondiente con un punto azul. Resolvemos para tiempos tales que nos permita ver el acercamiento en forma de espiral al punto de equilibrio. Los parámetros utilizados son $\beta = 0.15$, $\tau_{I} = 14$ y $\tau_{R} = 100$. 

```python
#funcion que calcula el punto de equilibrio pasando los parámetros
def eq_point(beta,tauI,tauR):
  s_eq = 1/(beta*tauI)
  i_eq = (beta*tauI-1)/(beta*(tauI+tauR))
  return s_eq,i_eq
```

```python
#grafico del retrato de fases adaptado de: https://kitchingroup.cheme.cmu.edu/blog/2013/02/21/Phase-portraits-of-a-system-of-ODEs/

def f(Y, t,beta,tauI,tauR):
    s, i = Y
    return [-beta*s*i + (1/tauR)*(1-s-i), beta*s*i - (1/tauI)*(i)]

y1 = np.linspace(0, 1.0, 15)
y2 = np.linspace(0, 1.0, 15)

Y1, Y2 = np.meshgrid(y1, y2)

t = 0

u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape

beta = 0.15
tauI = 14
tauR = 120

for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y], t,beta,tauI,tauR)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
     
#grafico el retrato de fase
Q = plt.quiver(Y1, Y2, u, v, color='r')

#grafico algunas trayectorias del sistema en particular
for i in range(10):
  # Total population, N.
  N = 10000
  # Initial number of infected and recovered individuals, I0 and R0.
  I0, R0 = (i*900+1)/N, 0
  # Everyone else, S0, is susceptible to infection initially.
  S0 = 1 - I0 - R0
  # Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
  beta, gamma = 0.15, 1./14
  # A grid of time points (in days)
  t = np.linspace(0, 1000, 1000)
  # Initial conditions vector
  y0 = S0, I0, R0
  # Integrate the SIR equations over the time grid, t.
  ret = odeint(deriv, y0, t, args=(beta, 1/gamma,120))
  S, I, R = ret.T
  plt.plot(S,I)

#grafico el punto de equilibrio
s_eq,i_eq = eq_point(beta,tauI,tauR)
print('El punto de equilibrio es: (',round(s_eq,2),',',round(i_eq,2),')')
plt.plot(s_eq,i_eq,'ob')

#detalles del gráfico
plt.xlabel('$s$ (suceptibles)',fontsize=16)
plt.ylabel('$i$ (infectados)',fontsize=16)
plt.xlim([0, 1.0])
plt.ylim([0, 1.0])
plt.show()
```

Adicionalmente, otra cosa que podemos hacer es resolver el sistema para valores fijos de parámetros y condiciones iniciales para observar directamente oscilaciones amortiguadas. En la siguiente celda, observamos la presencia de estas oscilaciones en un sistema en particular, en el cual introducimos inicialmente 1 infectado en una población de 10mil personas. Los parámetros son los mismos que al graficar el retrato de fase de la celda anterior. Se realiza la resolución un número de pasos que nos permita observar las oscilaciones que nos interesan. Con líneas de puntos, señalamos los puntos de equilibrios, como calculados teóricamente. 

```python
#ejemplo adaptado de la guía de SciPy: https://scipython.com/book/chapter-8-scipy/additional-examples/the-sir-epidemic-model/ 

# Total population, N.
N = 10000
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 1/N, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = 1 - I0 - R0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 0.15, 1./14
# A grid of time points (in days)
t = np.linspace(0, 1000, 1000)


# Initial conditions vector
y0 = S0, I0, R0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(beta, 1/gamma,120))
S, I, R = ret.T


# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(figsize=(10,8))
plt.plot(t, S, 'b', alpha=0.5, lw=3, label='Suceptibles')
plt.plot(t, I, 'r', alpha=0.5, lw=3, label='Infectados')
plt.plot(t, R, 'g', alpha=0.5, lw=3, label='Recuperados')

#grafico los puntos de equilibrio 
s_eq,i_eq = eq_point(beta,tauI,tauR)
r_eq = 1 - s_eq - i_eq
plt.plot(t,np.zeros(shape=t.shape)+s_eq,'--b',alpha=0.5)
plt.plot(t,np.zeros(shape=t.shape)+r_eq,'--g',alpha=0.5)
plt.plot(t,np.zeros(shape=t.shape)+i_eq,'--r',alpha=0.5)

plt.xlabel('Tiempo',fontsize=14)
plt.ylabel('$s,i,r$',fontsize=14)
plt.legend(fontsize=14)
plt.show()
```
