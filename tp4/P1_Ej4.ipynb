```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
plt.style.use('seaborn')
import scipy, scipy.stats
```

Primero, graficaremos cómo es la distribución de Poisson de media $\mu = 1.7$ para ver qué valores puede tomar la tasa de reproducción $r$. 

```python
mu = 1.7
x = np.arange(0,30,1)
pmf = scipy.stats.poisson.pmf(x,mu)
fig = plt.figure(figsize=(8,5))
plt.axvline(1.7,linestyle='--',label=r'$\lambda = 1.7$')
plt.plot(x,pmf)
plt.xlabel('$k$',fontsize=14)
plt.ylabel('$P(x=k)$',fontsize=14)
plt.legend(fontsize=14)
plt.show()
```

```python
def ej4(x0,pf):
    #xo: condicion inicial
    #pf: numero de iteraciones temporales
    x = []
    x.append(x0)
    r_vec = []
    i = 0
    while i < pf-1 and x[i] != 0: #paso temporal
        r = np.random.poisson(lam=1.7, size=1)
        r_vec.append(float(r))
        x.append(float(r*x[i]))
        i += 1
    return x,r_vec
```

Miremos primero algunas realizaciones posibles para ver los comportamientos que puede tener el sistema. 

```python
x,r = ej4(1,50)
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,5))
ax1.plot(x,'-ob')
ax1.set_xlabel(r'$t$',fontsize=22)
ax1.set_ylabel(r'$n_t$',fontsize=22)
ax2.plot(r,'-ob')
ax2.set_xlabel(r'$t$',fontsize=22)
ax2.set_ylabel(r'$r_t$',fontsize=22)
plt.show()
```

También, podemos encontrar una distribución de probabilidad para los tiempos o iteraciones a las cuales sucede la extinción de la especie. Para esto, corremos varias simulaciones y realizamos un histograma normalizado de estos números de iteraciones. 

```python
rep = 10000
t = []
for i in range(rep):
  x,r = ej4(1,50)
  if (x[-1] == 0.0):
    t.append(len(x)-1)
```

```python
mean = np.mean(t)
std = np.std(t)
print('Valor medio: ',round(mean,2))
print('Desviación Estándar: ',round(std,2))
plt.hist(t,density=True)
plt.xlabel(r'$t$',fontsize=18)
plt.ylabel(r'$t_{ext}$',fontsize=18)
plt.axvline(mean,linestyle='--',color='k',linewidth=3)
plt.show()
```
