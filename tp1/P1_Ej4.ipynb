```python
#@title 
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
plt.style.use('seaborn')
import scipy, scipy.stats
```

En este ejercicio, nos piden estudiar una especie con un ciclo de vida anual, en la cual un individuo produce un determinado número de descendientes y muere. Entonces, consideraremos que la población evoluciona según el sistema dinámico:

\begin{align}
n_{t+1} = r n_{t},
\end{align}

donde $r$ obedece una distribución de Poisson de media $\mu = 1.7$ y $N_0 = 1$. Es decir, comenzamos con un único individuo en la población. 

Primero, graficaremos cómo es la distribución de Poisson de media $\mu = 1.7$ para ver qué valores puede tomar la tasa de reproducción $r$. 

```python
#@title 
mu = 1.7
x = np.arange(0,30,1)
pmf = scipy.stats.poisson.pmf(x,mu)
fig = plt.figure(figsize=(8,5))
plt.axvline(1.7,linestyle='--',label=r'$\mu = 1.7$')
plt.axvline(1.7+np.sqrt(1.7),linestyle='--',color='k',label=r'$\mu + \sigma$')
plt.axvline(1.7-np.sqrt(1.7),linestyle='--',color='k',label=r'$\mu - \sigma$')
plt.plot(x,pmf)
plt.xlabel('$k$',fontsize=14)
plt.ylabel('$P(x=k)$',fontsize=14)
plt.legend(fontsize=14)
plt.show()
```

```python
#@title 
def ej4(x0,pf):
    #xo: condicion inicial
    #pf: numero de iteraciones temporales
    x = []
    x.append(x0)
    r_vec = []
    i = 0
    while i < pf-1 and x[i] != 0: #paso temporal
        r = np.random.poisson(lam=1.7, size=1)
        r_vec.append(float(r))
        x.append(float(r*x[i]))
        i += 1
    return x,r_vec
```

Primero, observemos algunas realizaciones posibles para ver los comportamientos que puede tener el sistema. Para ello, graficamos la población $n_t$ en función de las iteraciones. Adicionalmente, también graficamos los valores de $r$ que salen desde la primera iteración. 

Por un lado, observamos que la población eventualmente se extingue, cuando $r_t = 0$, para algún valor de $t$. Esto puede suceder tras unas pocas iteraciones, extinguiéndose rápidamente la población. Por otra parte, si obtenemos varios valores de $r$ no nulos sucesivamente, la población se mantiene o crece, antes de extinguirse. Para el último gráfico, la población llega a alrededor de $800$ individuos y se extingue después de 16 iteraciones. En cambio, para otra de las realizaciones, se compone como máximo de $8$ individuos y deja de existir en la iteración número 5. 

```python
#@title 
x,r = ej4(1,50)
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,5))
ax1.plot(x,'-ob')
ax1.set_xlabel(r'$t$',fontsize=22)
ax1.set_ylabel(r'$n_t$',fontsize=22)
ax2.plot(r,'-ob')
ax2.set_xlabel(r'$t$',fontsize=22)
ax2.set_ylabel(r'$r_t$',fontsize=22)
plt.show()
```

También, podemos encontrar una distribución de probabilidad para los tiempos o iteraciones a las cuales sucede la extinción de la especie. Para esto, corremos $10000$ simulaciones y realizamos un histograma normalizado de estos números de iteraciones de extinción, a los que llamaremos $t_{ext}$. Adicionalmente, calculamos un tiempo medio de extinción, y su desviación estándar. A estos los llamaremos $\mu^{*}$ y $\sigma^{*}$ para no generar confusión con los parámetros de la distribución de Poisson. 

Observamos que la probabilidad de obtener $0$ de muestrear una vez la distribución de Poisson es $e^{-1.7}$. Lo que estamos analizando en este histograma es la probabilidad de tener que muestrear $t_{ext}$ veces la variable aleatoria hasta obtener un $0$. En consecuencia, esta distribución de $t_{ext}$ debe ser una geométrica de parámetro $p = e^{-1.7}$ (inversa de la media). Por ello, junto con los valores de $\mu^{*}$ y $\sigma^{*}$ calculados, indicamos los de la distribución geométrica. Observamos que estos valores son similares. Finalmente, en línea continua verde, graficamos también la distribución geométrica de parámetro $p$. 

```python
#@title 
rep = 10000
t = []
for i in range(rep):
  x,r = ej4(1,50)
  if (x[-1] == 0.0):
    t.append(len(x)-1)
```

```python
#@title 
import scipy, scipy.stats
mean = np.mean(t)
std = np.std(t)
p = np.exp(-1.7)
meangeo = 1/p
stdgeo = np.sqrt(1-p)/p
print('Valor medio = ',round(mean,2),'(geométrica =',round(meangeo,2),')')
print('Desviación estándar = ',round(std,2),'(geométrica =',round(stdgeo,2),')')
plt.hist(t,density=True)
plt.xlabel(r'$t_{ext}$',fontsize=18)
plt.ylabel(r'$P(t = t_{ext})$',fontsize=18)
plt.axvline(mean,linestyle='--',color='r',linewidth=3,label=r'$\mu^{*}$')
plt.axvline(mean+std,linestyle='--',color='k',linewidth=3,label=r'$\mu^{*} + \sigma^{*}$')
plt.axvline(mean-std,linestyle='--',color='k',linewidth=3,label=r'$\mu^{*} - \sigma^{*}$')
plt.legend(fontsize=14)
x = np.linspace(0,40,41)
pmf = scipy.stats.geom.pmf(x,p)
plt.plot(x,pmf,'g',linewidth=3)
plt.show()
```
