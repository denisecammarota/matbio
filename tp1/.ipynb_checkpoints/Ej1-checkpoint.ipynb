```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#code for making coweb plot (source: https://www.johndcook.com/blog/2020/01/19/cobweb-plots/)
def cobweb_bevholt(f, x0, N, r, K,a=0, b=1):
    # plot the function being iterated
    t = np.linspace(a, b, N)
    plt.plot(t, f(t,r,K), 'k')

    # plot the dotted line y = x
    plt.plot(t, t, "k:")
    
    # plot the iterates
    x, y = x0, f(x0,r,K)
    for _ in range(N):
        fy = f(y,r,K)        
        plt.plot([x, y], [y,  y], 'b', linewidth=1)
        plt.plot([y, y], [y, fy], 'b', linewidth=1)
        x, y = y, fy
            
    plt.axes().set_aspect(1) 
    plt.xlabel(r'$n_t$',fontsize=15)
    plt.ylabel(r'$n_{t+1}$',fontsize=15)
    plt.show()
    plt.close()
```

```python
def bevholt(x,r,K):
    m = (r*x)/(1 + ((r-1)/K)*x)
    return m
```

```python
def sol_bevholt(x0,r,K,pf):
    #pf: numero de pasos final 
    #x0: condicion inicial
    sols = []
    sols.append(x0)
    for i in range(pf):
        x = (K*x0)/(x0 + (K-x0)*(r**(-(i+1))))
        sols.append(x)
    plt.plot(sols)
    plt.xlabel(r'$t$',fontsize=15)
    plt.ylabel(r'$n_t$',fontsize=15)
    plt.show()
```

# Hacemos algunos coweb plots para ver que sucede
# También graficamos la solución analítica del enunciado

# observamos los coweb plots para K > 0 y r > 1 

```python
cobweb_bevholt(bevholt, x0=0.2, N=20, r=2, K=1,a=0, b=1)
```

```python
sol_bevholt(x0=0.2, pf=20, r=2, K=1)
```

# observamos los coweb plots para K > 0 y r < 1 

```python
cobweb_bevholt(bevholt, x0=0.8, N=20, r=0.5, K=1,a=0, b=1)
```

```python
sol_bevholt(x0=0.8, pf=20, r=0.5, K=1)
```
