```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
plt.style.use('seaborn')
```

```python
def ej4(x0,pf):
    #xo: initial condition
    #pf: number of steps/iterations in time
    x = np.empty((pf,1))
    x[0] = x0
    r_vec = np.empty((pf-1,1))
    for i in range(pf-1):
        r = np.random.poisson(lam=1.7, size=1)
        r_vec[i] = np.copy(r)
        x[i+1] = r*x[i]
    print('Gráfico de n(t)')
    plt.xlabel(r'$t$',fontsize=15)
    plt.ylabel(r'$n_t$',fontsize=15)
    plt.plot(x)
    plt.show()
    print('Gráfico de r(t)')
    plt.plot(r_vec)
    plt.xlabel(r'$t$',fontsize=15)
    plt.ylabel(r'$n_t$',fontsize=15)
    plt.show()
```

```python
for i in range(10):
    print('*Ejecucción: ',i)
    ej4(1,20)
```
