El ejercicio 6 de esta práctica nos pide estudiar el **efecto Allee**, descripto por W.C. Allee, y que se asocia a la existencia de un número crítico mínimo de individuos para garantizar la supervivencia de una especie. Por debajo de un cierto umbral de población, el tamaño de la población es tan reducido que los individuos no se encuentran con otros de la misma especie y, por ende, no se reproducen. Una manera de modelar este efecto está dada por el sistema dinámico:

\begin{align}
\frac{dN}{dt} &= r N \Big( 1 - \frac{N}{K}\Big)\Big(\frac{N}{A} - 1\Big) \tag{1}. \\
\end{align}


## Equilibrios del sistema

Primero, nos piden analizar los equilibrios de este sistema. Simplemente, igualamos la primera ecuación a cero. Fácilmente, puede verse que esto se cumple únicamente para los puntos de equilibrio:


\begin{cases}
      n^{*}_{0} = 0  \\ 
      n^{*}_{1} = K \\
      n^{*}_{2} = A
  \end{cases}

Si recordamos el **mapeo logístico**, este tenía únicamente dos puntos de equilibrio correspondientes al origen y a $K$. En cambio, tenemos otro equilibrio para el modelo de Allee, que es $n^{*}_{2} = A$. Esto se debe a que el nuevo modelo tiene un término que multiplica a un mapeo logístico. 

## Análisis de estabilidad lineal

Ahora, analizaremos la estabilidad de los equilibrios realizando el análisis de estabilidad lineal. Para ello, definimos primero la función $f(N)$ como:

\begin{align}
f(N) &= r N \Big( 1 - \frac{N}{K}\Big)\Big(\frac{N}{A} - 1\Big), \\
\end{align}

cuya derivada está dada por la expresión:

\begin{align}
f'(N) &=  r \Big( 1 - \frac{N}{K}\Big)\Big(\frac{N}{A} - 1\Big) - r \frac{N}{K} \Big(\frac{N}{A} - 1\Big) - r \frac{N}{A} \Big( 1 - \frac{N}{K}\Big).
\end{align}

Evaluamos entonces la derivada en cada equilibrio, para hallar la estabilidad de los mismos en función de los parámetros del sistema. 

Primero, para $n^{*}_{0} = 0$, tenemos que:

\begin{align}
f'(0) &= r,
\end{align}

con lo cual el cero es estable si $r<0$, inestable si $r>0$ y marginalmente estable si $r = 0$

Para el segundo equilibrio, $n^{*}_{1} = K$, evaluamos la derivada y obtenemos que:

\begin{align}
f'(K) &= r \Big(\frac{K}{A} - 1\Big),
\end{align}

con lo cual la estabilidad depende del signo de esta derivada. Este está determinado por los valores de todos los parámetros del sistema, es decir, depende de $r,A$ y $K$.

Algo similar sucede para el tercer equilibrio $n^{*}_{2} = A$, ya que evaluando la derivada en este punto de equilibro, encontramos que:

\begin{align}
f'(A) &= r \big(\frac{A}{K} - 1\Big).
\end{align}

Es interesante observar que las derivadas de $f'(K)$ y $f'(A)$ tienen signos opuestos, lo cual sale de una simple inspección. 

En **resumen**, hemos hallado sobre las estabilidades de los equilibrios que:

\begin{align}
n^{*}_{0} = 0 &=
 \begin{cases}
      \text{estable}  & \text{si}\ r<0  \\ 
      \text{inestable}  & \text{si}\ r>0  \\
      \text{marginalmente estable}  & \text{si}\ r=0  \\
\end{cases}
\end{align}

\begin{align}
n^{*}_{1} = K &=
 \begin{cases}
      \text{estable}  & \text{si}\ r\Big(\frac{K}{A} - 1\Big) <0  \\ 
      \text{inestable}  & \text{si}\ r\Big(\frac{K}{A} - 1\Big) >0 \\
      \text{marginalmente estable}  & \text{si}\ r\Big(\frac{K}{A} - 1\Big) = 0 \\
\end{cases}
\end{align}

\begin{align}
n^{*}_{2} = A &=
 \begin{cases}

## Otros comentarios sobre estabilidad y diagramas de bifurcaciones

**Comentario 1**

Estamos trabajando con un sistema dinámico unidimensional. Por lo tanto, sabemos que la estabilidad de uno solo de estos equilibrios nos determina la de los equilibrios vecinos. De esta manera, conociendo la estabilidad de uno de ellos, encontramos de la del resto de los puntos equilibrio del sistema. Esto se ve claramente reflejado en lo que hallamos en la celda anterior. 

Veamos esto con un ejemplo en particular. 

Supongamos que $r>0$, y que $K>A$, de manera que $n^{*}_{0} < n^{*}_{2} < n^{*}_{1}$. Con la lógica de la alternancia de estabilidad de los equilibrios, tenemos que $n^{*}_{0}$ es inestable, con lo cual $n^{*}_{2}$ es estable y, por lo tanto, $n^{*}_{1}$ es inestable. Esto se reproduce también del cálculo matemático, ya que bajo estas hipótesis:

\begin{align}
f'(n^{*}_{0}) = f'(0) &= r < 0 \Rightarrow n^{*}_{0} \: \text{es inestable} \\
f'(n^{*}_{2}) = f'(A) &= r\Big(\frac{A}{K} - 1\Big) < 0 \Rightarrow n^{*}_{2} \: \text{es estable} \\
f'(n^{*}_{1}) = f'(K) &= r \Big(\frac{K}{A} - 1\Big) < 0 \Rightarrow n^{*}_{1} \: \text{es inestable}
\end{align}

Es posible fácilmente realizar este análisis para otros valores de los parámetros.
